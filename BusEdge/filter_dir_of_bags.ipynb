Helper notebook that filters a directory of rosbag files using user-provided GPS coordinates.

```python
import glob
import time
```

```python
DIR_OF_BAGS = 'path/to/directory/of/BusEdge/bags/*'
```

```python
all_bag_names = glob.glob(DIR_OF_BAGS)
all_bag_names.sort()
```

```python
print('Iterating through bags...')
start = time.time()
for bag in all_bag_names:
    print('\tBag', bag)
    dump_folder = bag.split('/')[-1][:-4] + '_' + 'dump'
#     !python BusEdge/client/scripts/use_rosbag/filter_bag.py -i placeholder.bag -c 3 -o $dump_folder --gps-filtering 40.17848333333333 -80.24144666666666 40.175305 -80.243945 # stretch 1
    !python BusEdge/client/scripts/use_rosbag/filter_bag.py -i $bag -c 3 -o $dump_folder --gps-filtering 40.17544 -80.24375 40.17435666666667 -80.24526333333333 # stretch 2
print('\nTook %f minutes to execute' % ((time.time()-start)/60))
```
